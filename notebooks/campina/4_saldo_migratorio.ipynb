```python
from br_demography  import municipality_migration as mm
import basedosdados as bd
from dotenv import load_dotenv
import os
import pandas as pd

# Carrega as variáveis de ambiente do arquivo .env
load_dotenv(dotenv_path='./br_demography/.env')

# Use as variáveis de ambiente conforme necessário
project_id = os.getenv('GOOGLE_CLOUD_PROJECT_ID')
```

```python
mm.query_total_population(mun_id=4104006, project_id=project_id)
```

```python
df_emigration = mm.query_emigration_by_sex_age(mun_id=4104006, project_id=project_id)
df_emigration
```

```python
df_emigration = mm.standard_age_groups(df=df_emigration, age_group_csv_path='./br_demography/source/tab/faixas_etarias_censo_2000_2010.csv')
df_emigration
```

```python
df_imigration = mm.query_imigration_by_sex_age(mun_id=4104006, project_id=project_id)
df_imigration
```

```python
df_imigration = mm.standard_age_groups(df=df_imigration, age_group_csv_path='./br_demography/source/tab/faixas_etarias_censo_2000_2010.csv')
df_imigration
```

```python
df_saldo = ((df_imigration - df_emigration) /5).round().astype(int)
```

```python
df_saldo.to_csv(path_or_buf='./br_demography/results/tab/saldo_migratorio_anual_2000_2010_campina_2022_2044.csv', sep=';', decimal=',', encoding='utf-16')
```
